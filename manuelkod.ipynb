```python
import pandas as pd

# DataFrame oluşturma
df = pd.read_csv('cevreselhaberler02.csv')
df = df[0:300]
df
```

```python

# Yeni kolon için boş bir liste oluşturma
cevre_category = []

# Her bir satır için kullanıcıdan girdi alma
for index, row in df.iterrows():
    print(f"Row {index}: {row.to_dict()}")
    value = input(f"Please enter a value (0 or 1) for Age_Category for Row {index}: ")
    while value not in ["0", "1"]:
        value = input("Invalid input! Please enter 0 or 1: ")
    cevre_category.append(int(value))

# Yeni kolonu DataFrame'e ekleme
df['cevre_category'] = cevre_category

# Sonuçları görüntüleme
print(df)

```

```python
df.to_csv("çevehaberlerikatogorizeedilmis.csv")
```
